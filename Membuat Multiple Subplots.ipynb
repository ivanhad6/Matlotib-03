```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Simple Line Plot

x=np.arange(0.0,2.0,0.01)
x
```

```python
s=np.sin(2*np.pi*x)
s
```

```python
fig, ax=plt.subplots() #OO Style
ax.plot(x, s)
ax.set(xlabel='nilai x', ylabel='nilai sin(x)', title='Visualisasi nilai sin')
ax.grid()
plt.show()
```

```python
#Multiple Subplots
x1= np.linspace(0.0,5.0,100)
x2= np.linspace(0.0,2.0,100)

y1=np.cos(2*np.pi*x1)
y2=np.cos(2*np.pi*x2)
```

```python
x1
```

```python
#Multiple Subplots Dengan  Menggunakan OO Style

fig, (ax1,ax2)=plt.subplots(2,1) #menggunakan 2 parameter dengan komposisi 2,1 , 2 jumlah baris, dan 1 kolom
fig.suptitle('Multiple Subplots')  #untuk menyertakan judul 

ax1.plot(x1,y1,'ro-') #r berarti red, o berarti bentuk marker, - adalah garis
ax1.set_ylabel('Nilai $cos(x_1)$') #untuk memberi nama label #simbol $ berfungsi untuk memberi spasi pada label

ax2.plot(x2,y2,'g.-')
ax2.set_ylabel('Nilai $cos(x_2)$')

ax2.set_xlabel('Nilai x')
```

```python
#Multiple Subplots Dengan pyplot Style

plt.subplot(2,1,1) #parameter pertama berfungsi untuk mengatur layout dan parameter ke-3 berfungsi untuk mengendalikan axes pertama.
plt.plot(x1,y1,'ro-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai $cos(x_1)$')

plt.subplot(2,1,2) #parameter pertama berfungsi untuk mengatur layout dan parameter ke-3 berfungsi untuk mengendalikan axes kedua
plt.plot(x2,y2,'g.-')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai $x$')

plt.show()
```
